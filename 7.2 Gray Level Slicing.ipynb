```python
#Mengimport Library
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```python
#Membaca gambar dari file yang disimpan dengan cv2
img = cv2.imread('prambanan.jpg', cv2.IMREAD_GRAYSCALE)

#Mendapatkan dimensi baris dan kolom dari gambar
row, column = img.shape

#Membuat citra baru dengan ukuran yang sama seperti citra asli untuk menginisialisasi matriks nol dengan tipe data
img1 = np.zeros((row,column),dtype = 'uint8')
 
#Menentukan rentang ambang batas minimum dan maksimum
min_range = 10
max_range = 60
 
#Melakukan loop pada setiap piksel dalam citra img dan menghasilkan citra biner baru img1 berdasarkan ambang batas minimum dan maksimum yang ditentukan
for i in range(row):
    for j in range(column):
        if img[i,j]>min_range and img[i,j]<max_range:
            img1[i,j] = 255
        else:
            img1[i,j] = 0
```

```python
#Plot Image
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
ax = axes.ravel()

ax[0].imshow(img, cmap=plt.cm.gray)
ax[0].set_title("Citra Input")
ax[1].hist(img.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(img1, cmap=plt.cm.gray)
ax[2].set_title("Citra Output")
ax[3].hist(img1.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```
